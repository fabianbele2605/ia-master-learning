```python
# 🔹 CELDA 2: Vector DB Simulado

print("\n" + "=" * 60)
print("VECTOR DATABASE: Recuperación de documentos")
print("=" * 60)

print("""
¿QUÉ HACE ESTA CELDA?
Implementa búsqueda de documentos por similitud.
(En prod: Pinecone, Chroma, Weaviate)

IDEA:
1. Documentos → Embeddings (vectores)
2. Pregunta → Embedding
3. Buscar docs más cercanos (similitud coseno)
4. Pasar a LLM
""")

# Documents base de conocimiento
documents = {
    1: "Albert Einstein fue un físico teórico nacido en 1879. Revolucionó la física con la teoría de la relatividad.",
    2: "Isaac Newton nació en 1643 e inventó el cálculo. Sus leyes de movimiento son fundamentales.",
    3: "Marie Curie fue una física y química polaca. Descubrió el polonio y el radio.",
    4: "Python es un lenguaje de programación versátil. Usado para ML, web, ciencia de datos.",
    5: "Machine Learning es un subcampo de IA. Los modelos aprenden patrones de datos.",
}

# Simular embeddings (en prod: use transformers library)
np.random.seed(42)
doc_embeddings = {doc_id: np.random.randn(8) for doc_id in documents}

def similarity(vec1, vec2):
    """Similitud coseno"""
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

def retrieve_docs(query, doc_embeddings, documents, top_k=2):
    """
    ¿QUÉ HACE?
    1. Convierte query a embedding
    2. Calcula similitud con todos los docs
    3. Retorna top-k más similares
    """
    # Simular embedding de query
    query_embedding = np.random.randn(8)
    
    # Calcular similitudes
    scores = {}
    for doc_id, doc_emb in doc_embeddings.items():
        scores[doc_id] = similarity(query_embedding, doc_emb)
    
    # Top-k
    top_docs = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:top_k]
    
    return top_docs

# Test
query = "¿Quién inventó el cálculo?"
top_docs = retrieve_docs(query, doc_embeddings, documents, top_k=2)

print(f"\n🔍 Query: \"{query}\"")
print(f"\n📄 Documentos recuperados:")
for i, (doc_id, score) in enumerate(top_docs):
    print(f"\n  {i+1}. (similitud: {score:.3f})")
    print(f"     \"{documents[doc_id][:80]}...\"")

print(f"\n✅ Vector DB funciona:")
print(f"  - Documentos como vectores")
print(f"  - Búsqueda rápida por similitud")
print(f"  - Top-K recuperación")
```

```python
# 🔹 CELDA 3: RAG Pipeline

print("\n" + "=" * 60)
print("RAG PIPELINE COMPLETO")
print("=" * 60)

print("""
¿QUÉ HACE ESTA CELDA?
Implementa RAG end-to-end:
Pregunta → Retrieve docs → LLM genera respuesta
""")

class RAGSystem:
    """
    ¿QUÉ HACE?
    1. Recupera documentos relevantes
    2. Construye contexto
    3. LLM (simulado) genera respuesta
    """
    
    def __init__(self, documents, doc_embeddings):
        self.documents = documents
        self.doc_embeddings = doc_embeddings
    
    def retrieve(self, query, top_k=2):
        """Recuperar documentos"""
        return retrieve_docs(query, self.doc_embeddings, self.documents, top_k)
    
    def generate(self, query, retrieved_docs):
        """
        ¿QUÉ HACE?
        Simula LLM que lee documentos y responde.
        (En prod: LLM real como GPT-3, Claude)
        """
        # Construir contexto
        context = "\n".join([
            f"Doc: {self.documents[doc_id]}" 
            for doc_id, _ in retrieved_docs
        ])
        
        # Simular respuesta LLM
        if "Einstein" in query or "relatividad" in documents.get(retrieved_docs[0][0], ""):
            answer = "Albert Einstein fue un físico teórico nacido en 1879. Es famoso por la teoría de la relatividad que revolucionó la física."
        elif "Newton" in query or "cálculo" in documents.get(retrieved_docs[0][0], ""):
            answer = "Isaac Newton fue un físico y matemático que nació en 1643. Inventó el cálculo e formuló las leyes del movimiento."
        else:
            answer = "Basándome en los documentos recuperados: " + retrieved_docs[0][0] if retrieved_docs else "No encontré información."
        
        return answer
    
    def answer(self, query):
        """RAG completo"""
        retrieved = self.retrieve(query)
        answer = self.generate(query, retrieved)
        return answer, retrieved

# Usar RAG
rag = RAGSystem(documents, doc_embeddings)

queries = [
    "¿Quién inventó el cálculo?",
    "¿Cuándo nació Albert Einstein?",
    "¿Qué es Machine Learning?",
]

print(f"\n🤖 RAG System en acción:\n")

for query in queries:
    answer, retrieved = rag.answer(query)
    print(f"❓ Q: {query}")
    print(f"📚 Docs usados: {len(retrieved)}")
    print(f"💬 A: {answer}\n")

print(f"✅ RAG funcionando:")
print(f"  - Recupera documentos relevantes")
print(f"  - LLM genera respuesta contextual")
print(f"  - Respuestas precisas y trazables")
```
