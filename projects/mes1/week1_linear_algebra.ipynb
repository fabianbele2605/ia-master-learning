```python
# Celda 1: Importar librerias
import numpy as np
import matplotlib.pyplot as plt

print(" Libreria importadas")
```

```python
# Celda 2: Crear matrices
# Crear una matriz 2x2
A= np.array([[1, 2],
             [3, 4]])

print("Matriz A:")
print(A)
print(f"Shape: {A.shape}") # (2, 2)
```

```python
# Celda 3: Operaciones basicas
# Transpuesta
print("Transpuesta de A:")
print(A.T)

# Suma 
B= np.array([[5, 6], [7, 8]])
print("\nA + B:")
print(A + B)

# Multiplicacion elemento a elemento
print("\nA * B (elemento a elemento):")
print(A * B)

# Multiplicacion matricial
print("\nA @ B (producto matricial):")
print(A @ B)
```

```python
# Celda 4: Determinante e inversa
# Determinate 
det_A = np.linalg.det(A)
print(f"Determinate de A: {det_A}")

# Inversa
A_inv = np.linalg.inv(A)
print("\nInversa de A:")
print(A_inv)

# Verifica: A @ A_inv = I (matriz identidad)
print("\nA @ A_inv (deberia ser aproximadamente [[1, 0], [0, 1]]):")
print(A @ A_inv)
```

```python
# Celda 5: Eigenvalues y Eigenvectors
# Calcular eigenvalues y eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)
```

```python
# Celda 6: Visualización
#Visualiza eigenvectors como flechas
fig, ax = plt.subplots(figsize=(8, 8))

# Dibuja los eigenvectors
for i in range(2):
    vec = eigenvectors[:, i]
    ax.arrow(0, 0, vec[0], vec[1], head_width=0.1, head_length=0.1, fc=f'C{i}', ec=f'C{i}')
    ax.text(vec[0]*1.15, vec[1]*1.15, f'EV{i+1}\n(λ={eigenvalues[i]:.2f})', fontsize=12)

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.set_title('Eigenvectors de A', fontsize=14)
plt.show()
```

```python
# Celda 7: SVD (Singular Value Decomposition)
# SVD: A = U @ Σ @ V^T
U, s, Vt = np.linalg.svd(A)

print("U (matriz ortogonal):")
print(U)

print("\nSingular values (Σ):")
print(s)

print("\nV^T (matriz ortogonal transpuesta):")
print(Vt)

# Reconstruye A
S = np.diag(s)
A_reconstructed = U @ S @ Vt
print("\nA reconstruida (U @ Σ @ V^T):")
print(A_reconstructed)
print("Original A:")
print(A)
```

```python
### SVD Manual

print("=" * 60)
print("SVD DESDE CERO (Sin usar np.linalg.svd)")
print("=" * 60)

import numpy as np

# Matriz A
A = np.array([[1, 2], [3, 4]], dtype=float)
print("\nMatriz A:")
print(A)

# PASO 1: Calcular A^T @ A
AtA = A.T @ A
print("\nPaso 1: A^T @ A")
print(AtA)

# PASO 2: Eigenvalues y eigenvectors de A^T @ A
eigenvalues, eigenvectors = np.linalg.eig(AtA)
print("\nPaso 2: Eigenvalues (σ²) y eigenvectors")
print(f"Eigenvalues: {eigenvalues}")

# PASO 3: Singular values (Raiz de eigenvalues)
sigma = np.sqrt(eigenvalues)
print(f"\nPaso 3: Singular values (σ) = √(eigenvalues)")
print(f"σ = {sigma}")

# PASO 4: Ordenar por magnitud (mayor primero)
idx = np.argsort(-sigma)  # Descendente
sigma = sigma[idx]
V = eigenvectors[:, idx]

print(f"\nPaso 4: Reordenar manigtud")
print(f"σ (ordenado): {sigma}")
print (f"V (eigenvectors):\n{V}")

# PASO 5: Calcular U
U = np.zeros_like(A, dtype=float)
for i in range(len(sigma)):
    if sigma[i] > 1e-10:  # Evita división por 0
        U[:, i] = (A @ V[:, i]) / sigma[i]

print(f"\nPaso 5: U = (A @ V) / σ")
print(f"U:\n{U}")

# VERIFICACIÓN: Reconstruye A
S = np.diag(sigma)
A_reconstructed = U @ S @ V.T

print(f"\n✅ VERIFICACIÓN: A = U @ Σ @ V^T")
print(f"\nA original:\n{A}")
print(f"\nA reconstruida:\n{A_reconstructed}")
print(f"\nDiferencia (error):\n{A - A_reconstructed}")
print(f"\nError máximo: {np.max(np.abs(A - A_reconstructed)):.2e}")
```

```python
# 🧪 Paso 2: Experimenta con Diferentes Matrices

print("\n" + "=" * 60)
print("EXPERIMENTA: ¿Qué significan σ, U, V?")
print("=" * 60)

# Matriz 1: Solo escala
print("\n1️⃣ MATRIZ QUE SOLO ESCALA (2x en x, 3x en y)")
A1 = np.array([[2, 0], [0, 3]], dtype=float)
print("A1:")
print(A1)

# Usa SVD de numpy para ver resultado
U1, s1, Vt1 = np.linalg.svd(A1)
print(f"Singular values: {s1}")
print(f"Interpretación: Escala 2x y 3x ✓")

# Matriz 2: Rotación 45 grados
print("\n2️⃣ MATRIZ QUE ROTA 45 GRADOS")
theta = np.pi / 4
A2 = np.array([[np.cos(theta), -np.sin(theta)],
               [np.sin(theta), np.cos(theta)]])
print("A2:")
print(A2)

U2, s2, Vt2 = np.linalg.svd(A2)
print(f"Singular values: {s2}")
print(f"Interpretación: Solo rota, no escala (σ ≈ 1) ✓")

# Matriz 3: Combinación (escala + rotación)
print("\n3️⃣ MATRIZ COMBINADA (escala + rotación)")
A3 = np.array([[2, 1], [1, 2]], dtype=float)
print("A3:")
print(A3)

U3, s3, Vt3 = np.linalg.svd(A3)
print(f"Singular values: {s3}")
print(f"U (rotación inicial):\n{U3}")
print(f"V (rotación final):\n{Vt3.T}")
print(f"σ (escala): {s3}")
```

```python
# 📊 Paso 3: Visualización Avanzada

import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Función auxiliar
def plot_transformation(ax, A, title):
    # Vectores base originales
    v1 = np.array([1, 0])
    v2 = np.array([0, 1])
    
    # Transformados
    v1_transformed = A @ v1
    v2_transformed = A @ v2
    
    # Plot
    ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.006, label='e1 original')
    ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.006, label='e2 original')
    
    ax.quiver(0, 0, v1_transformed[0], v1_transformed[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.006, label='A@e1')
    ax.quiver(0, 0, v2_transformed[0], v2_transformed[1], angles='xy', scale_units='xy', scale=1, color='orange', width=0.006, label='A@e2')
    
    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.set_title(title, fontsize=12)
    ax.legend(fontsize=8)

# 3 matrices diferentes
A_escala = np.array([[2, 0], [0, 3]])
A_rotacion = np.array([[0, -1], [1, 0]])  # 90 grados
A_general = np.array([[1, 2], [3, 4]])

plot_transformation(axes[0], A_escala, "Escala (2x, 3x)")
plot_transformation(axes[1], A_rotacion, "Rotación 90°")
plot_transformation(axes[2], A_general, "Combinada")

plt.tight_layout()
plt.show()

print("Observa cómo cada matriz transforma los vectores base")
```
