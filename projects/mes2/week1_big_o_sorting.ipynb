## Celda 1: Big O Notation - La Teoría

**¿QUÉ HACE ESTA CELDA?**
Define Big O notation: cómo crece el tiempo de ejecución con el tamaño de entrada.

```
O(1)      = Tiempo constante (hash lookup)
O(log n)  = Tiempo logarítmico (binary search)
O(n)      = Tiempo lineal (for loop)
O(n log n)= Tiempo lineal-logarítmico (merge sort)
O(n²)     = Tiempo cuadrático (bubble sort)
O(2^n)    = Tiempo exponencial (brute force)
O(n!)     = Factorial (todas las permutaciones)
```

````python
import numpy as np
import matplotlib.pyplot as plt
import time
from typing import List, Callable

# Visualizar cómo crece cada complejidad
n_values = np.arange(1, 101)

complexities = {
    'O(1)': np.ones_like(n_values),
    'O(log n)': np.log(n_values),
    'O(n)': n_values,
    'O(n log n)': n_values * np.log(n_values),
    'O(n²)': n_values ** 2,
    'O(2^n)': 2 ** (n_values / 20),  # Scaled down para visualizar
}

plt.figure(figsize=(12, 6))
for label, values in complexities.items():
    plt.plot(n_values, values, label=label, linewidth=2)

plt.xlabel('Input Size (n)', fontsize=12)
plt.ylabel('Operations', fontsize=12)
plt.title('Big O Complexity Growth Rates', fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(alpha=0.3)
plt.yscale('log')
plt.tight_layout()
plt.show()

print("Comparación para n=100:")
print(f"O(1):      1 operación")
print(f"O(log n):  {np.log(100):.1f} operaciones")
print(f"O(n):      100 operaciones")
print(f"O(n log n): {100 * np.log(100):.0f} operaciones")
print(f"O(n²):     10,000 operaciones")
````

## Celda 2: Bubble Sort - O(n²)

**¿QUÉ HACE ESTA CELDA?**
Implementar bubble sort (el más simple pero lento). Compara elementos adyacentes repetidamente.

````python
def bubble_sort(arr: List[int]) -> List[int]:
    """Bubble sort: compara elementos adyacentes y los intercambia si están desordenados."""
    arr = arr.copy()
    n = len(arr)
    
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        
        # Si no hubo swaps, el array está ordenado
        if not swapped:
            break
    
    return arr

# Test
test = [64, 34, 25, 12, 22, 11, 90]
print(f"Original: {test}")
print(f"Bubble Sort: {bubble_sort(test)}")

# Análisis:
print("\nAnálisis Bubble Sort:")
print(f"Mejor caso (ya ordenado): O(n) - una pasada")
print(f"Promedio: O(n²)")
print(f"Peor caso (invertido): O(n²) - todas las pasadas")
print(f"Espacio: O(1) - in-place sorting")
````

## Celda 3: Quick Sort - O(n log n) promedio

**¿QUÉ HACE ESTA CELDA?**
Implementar quicksort usando divide-and-conquer. Elige un pivot y particiona alrededor de él.

````python
def quick_sort(arr: List[int]) -> List[int]:
    """Quick sort: divide-and-conquer usando pivot."""
    if len(arr) <= 1:
        return arr
    
    pivot = arr[len(arr) // 2]  # Elemento del medio
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    
    return quick_sort(left) + middle + quick_sort(right)

# Test
test = [64, 34, 25, 12, 22, 11, 90]
print(f"Original: {test}")
print(f"Quick Sort: {quick_sort(test)}")

# Análisis
print("\nAnálisis Quick Sort:")
print(f"Mejor caso: O(n log n) - pivot siempre en medio")
print(f"Promedio: O(n log n) - muy rápido en práctica")
print(f"Peor caso: O(n²) - pivot siempre en extremo")
print(f"Espacio: O(log n) - recursión stack")
````

## Celda 4: Merge Sort - O(n log n) garantizado

**¿QUÉ HACE ESTA CELDA?**
Implementar mergesort. Divide el array a la mitad recursivamente y luego mezcla los mitades ordenadas.

````python
def merge_sort(arr: List[int]) -> List[int]:
    """Merge sort: divide-and-conquer garantizado O(n log n)."""
    if len(arr) <= 1:
        return arr
    
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    
    return merge(left, right)

def merge(left: List[int], right: List[int]) -> List[int]:
    """Mezcla dos arrays ordenados."""
    result = []
    i = j = 0
    
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    result.extend(left[i:])
    result.extend(right[j:])
    return result

# Test
test = [64, 34, 25, 12, 22, 11, 90]
print(f"Original: {test}")
print(f"Merge Sort: {merge_sort(test)}")

# Análisis
print("\nAnálisis Merge Sort:")
print(f"Mejor caso: O(n log n)")
print(f"Promedio: O(n log n)")
print(f"Peor caso: O(n log n) - GARANTIZADO")
print(f"Espacio: O(n) - crea arrays temporales")
print(f"Ventaja: Estable (mantiene orden de elementos iguales)")
````

## Celda 5: Heap Sort - O(n log n) con espacio O(1)

**¿QUÉ HACE ESTA CELDA?**
Implementar heapsort. Usa estructura heap (árbol binario completo) para ordenar in-place.

````python
def heap_sort(arr: List[int]) -> List[int]:
    """Heap sort: O(n log n) in-place usando estructura heap."""
    arr = arr.copy()
    n = len(arr)
    
    # Construir max-heap
    for i in range(n // 2 - 1, -1, -1):
        heapify_down(arr, i, n)
    
    # Extraer elementos del heap
    for i in range(n - 1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]
        heapify_down(arr, 0, i)
    
    return arr

def heapify_down(arr: List[int], i: int, n: int):
    """Mantiene la propiedad max-heap."""
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2
    
    if left < n and arr[left] > arr[largest]:
        largest = left
    if right < n and arr[right] > arr[largest]:
        largest = right
    
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify_down(arr, largest, n)

# Test
test = [64, 34, 25, 12, 22, 11, 90]
print(f"Original: {test}")
print(f"Heap Sort: {heap_sort(test)}")

# Análisis
print("\nAnálisis Heap Sort:")
print(f"Mejor/Promedio/Peor: O(n log n)")
print(f"Espacio: O(1) - in-place")
print(f"Ventaja: O(n log n) garantizado + O(1) espacio")
print(f"Desventaja: No es estable (pierde orden de iguales)")
````

## Celda 6: Insertion Sort - O(n²) pero rápido en práctica

**¿QUÉ HACE ESTA CELDA?**
Implementar insertion sort. Inserta cada elemento en su posición correcta en la parte ordenada.

````python
def insertion_sort(arr: List[int]) -> List[int]:
    """Insertion sort: O(n²) pero rápido para arrays pequeños."""
    arr = arr.copy()
    
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        
        arr[j + 1] = key
    
    return arr

# Test
test = [64, 34, 25, 12, 22, 11, 90]
print(f"Original: {test}")
print(f"Insertion Sort: {insertion_sort(test)}")

# Análisis
print("\nAnálisis Insertion Sort:")
print(f"Mejor caso: O(n) - ya ordenado")
print(f"Promedio: O(n²)")
print(f"Peor caso: O(n²)")
print(f"Espacio: O(1) - in-place")
print(f"Ventaja: Estable, rápido para arrays pequeños")
print(f"Uso real: Python sorted() usa Timsort (insertion + merge para pequeños)")
````

## Celda 7: Benchmark - Comparar rendimiento real

**¿QUÉ HACE ESTA CELDA?**
Medir el tiempo real de cada algoritmo en arrays de diferentes tamaños. Verificar que la teoría (Big O) coincide con la práctica.

````python
def benchmark_sort(sort_func: Callable, arr: List[int], name: str) -> float:
    """Mide el tiempo de ejecución de un algoritmo de sorting."""
    start = time.time()
    sort_func(arr.copy())
    elapsed = time.time() - start
    return elapsed

# Crear arrays de prueba
sizes = [100, 500, 1000, 2000, 5000]
results = {name: [] for name in ['Bubble Sort', 'Insertion Sort', 'Quick Sort', 
                                   'Merge Sort', 'Heap Sort', 'Python sort()']}

print("Benchmarking algorithms...")
for size in sizes:
    arr = np.random.randint(0, 10000, size).tolist()
    
    # Bubble sort muy lento para arrays grandes
    if size <= 1000:
        results['Bubble Sort'].append(benchmark_sort(bubble_sort, arr, 'bubble'))
    else:
        results['Bubble Sort'].append(None)
    
    # Insertion sort también lento
    if size <= 2000:
        results['Insertion Sort'].append(benchmark_sort(insertion_sort, arr, 'insertion'))
    else:
        results['Insertion Sort'].append(None)
    
    results['Quick Sort'].append(benchmark_sort(quick_sort, arr, 'quick'))
    results['Merge Sort'].append(benchmark_sort(merge_sort, arr, 'merge'))
    results['Heap Sort'].append(benchmark_sort(heap_sort, arr, 'heap'))
    results['Python sort()'].append(benchmark_sort(lambda x: sorted(x), arr, 'python'))

# Visualizar resultados
plt.figure(figsize=(14, 6))

for name, times in results.items():
    valid_sizes = [s for i, s in enumerate(sizes) if times[i] is not None]
    valid_times = [t for t in times if t is not None]
    if valid_times:
        plt.plot(valid_sizes, valid_times, marker='o', label=name, linewidth=2)

plt.xlabel('Array Size (n)', fontsize=12)
plt.ylabel('Time (seconds)', fontsize=12)
plt.title('Sorting Algorithm Performance Comparison', fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print("\nResultados (segundos):")
print(f"{'Size':<10} {results.keys()}")
for size, idx in zip(sizes, range(len(sizes))):
    print(f"{size:<10}", end='')
    for name in results:
        time_val = results[name][idx]
        if time_val:
            print(f"{time_val:.4f}s", end=" ")
        else:
            print("  N/A  ", end=" ")
    print()
````

## Celda 8: LeetCode Practice - Sorting Problems

**¿QUÉ HACE ESTA CELDA?**
Resolver 3 problemas clásicos de sorting para practicar razonamiento algorítmico.

````python
# Problema 1: Merge Sorted Array (LeetCode 88)
def merge_sorted_arrays(nums1: List[int], m: int, nums2: List[int], n: int) -> None:
    """
    Mezcla nums2 en nums1 (que tiene espacio suficiente).
    Do it in-place: O(m+n) tiempo, O(1) espacio.
    """
    p1 = m - 1
    p2 = n - 1
    p = m + n - 1
    
    while p1 >= 0 and p2 >= 0:
        if nums1[p1] > nums2[p2]:
            nums1[p] = nums1[p1]
            p1 -= 1
        else:
            nums1[p] = nums2[p2]
            p2 -= 1
        p -= 1
    
    # Si quedan elementos en nums2, copiarlos
    while p2 >= 0:
        nums1[p] = nums2[p2]
        p2 -= 1
        p -= 1

# Test
nums1 = [1, 2, 3, 0, 0, 0]
merge_sorted_arrays(nums1, 3, [2, 5, 6], 3)
print(f"Merge Sorted Array: {nums1}")
assert nums1 == [1, 2, 2, 3, 5, 6], "Fallo en merge"
print("✓ Passed")

# Problema 2: Find Kth Largest (LeetCode 215)
def find_kth_largest(nums: List[int], k: int) -> int:
    """
    Encuentra el k-ésimo elemento más grande.
    Usar quickselect: O(n) promedio, O(n²) peor caso.
    """
    def quickselect(left, right, k_smallest):
        if left == right:
            return nums[left]
        
        # Partición
        pivot_idx = np.random.randint(left, right + 1)
        pivot_idx = partition(left, right, pivot_idx)
        
        if k_smallest == pivot_idx:
            return nums[k_smallest]
        elif k_smallest < pivot_idx:
            return quickselect(left, pivot_idx - 1, k_smallest)
        else:
            return quickselect(pivot_idx + 1, right, k_smallest)
    
    def partition(left, right, pivot_idx):
        pivot = nums[pivot_idx]
        nums[pivot_idx], nums[right] = nums[right], nums[pivot_idx]
        store_idx = left
        
        for i in range(left, right):
            if nums[i] < pivot:
                nums[store_idx], nums[i] = nums[i], nums[store_idx]
                store_idx += 1
        
        nums[right], nums[store_idx] = nums[store_idx], nums[right]
        return store_idx
    
    return quickselect(0, len(nums) - 1, len(nums) - k)

# Test
nums = [3, 2, 1, 5, 6, 4]
result = find_kth_largest(nums.copy(), 2)
print(f"Kth Largest (k=2): {result}")
assert result == 5, f"Expected 5, got {result}"
print("✓ Passed")

# Problema 3: Sort Colors (LeetCode 75)
def sort_colors(nums: List[int]) -> None:
    """
    Ordena array de 0s, 1s, 2s (colores) in-place.
    O(n) tiempo, O(1) espacio usando 3-way partición.
    """
    p0 = p2 = 0
    curr = 0
    
    while curr <= p2:
        if nums[curr] == 0:
            nums[p0], nums[curr] = nums[curr], nums[p0]
            p0 += 1
            curr += 1
        elif nums[curr] == 2:
            nums[curr], nums[p2] = nums[p2], nums[curr]
            p2 -= 1
        else:
            curr += 1

# Test
nums = [2, 0, 2, 1, 1, 0]
sort_colors(nums)
print(f"Sort Colors: {nums}")
assert nums == [0, 0, 1, 1, 2, 2], "Fallo en sort colors"
print("✓ Passed")
````
