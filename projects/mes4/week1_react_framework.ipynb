```python
# CELDA 1: ReAct Framework Introduction

print("=" * 70)
print("MES 4, WEEK 1: ReAct - Reasoning + Acting Framework")
print("=" * 70)

intro = """
QUE ES UN AGENTE AUTONOMO?

Programa que:
  1. Recibe objetivo (task/query)
  2. Razona que hacer (Thought)
  3. Ejecuta accion (Action)
  4. Observa resultado (Observation)
  5. Decide siguiente paso
  6. Repite hasta completar objetivo

DIFERENCIA: Agent vs Script

SCRIPT: input -> process -> output (linear)
AGENT: input -> think -> act -> observe -> ... (iterativo)
"""

print(intro)
```

```python
# CELDA 2: ReAct Loop

print("\n" + "=" * 70)
print("ReAct LOOP")
print("=" * 70)

react_loop = """
INPUT: Cual es la capital de Espana?
       |
       V
THOUGHT: Necesito web_search
         |
         V
ACTION: web_search(capital de Espana)
        |
        V
OBSERVATION: Madrid es la capital
             |
             V
             Tengo respuesta? SI
             |
             V
FINAL ANSWER: Madrid
"""

print(react_loop)
```

```python
# CELDA 4: Pseudocodigo

print("\n" + "=" * 70)
print("PSEUDOCODIGO: ReAct Agent")
print("=" * 70)

pseudocode = """
function ReActAgent(query, tools, max_iterations=10):
    history = []
    
    for i in range(max_iterations):
        # 1. REASONING: Ask LLM
        prompt = format_prompt(query, history)
        llm_output = llm.generate(prompt)  # <- LLM thinks
        
        # 2. PARSE: Extract thought, action, answer
        thought, action, action_input = parse(llm_output)
        history.append({thought: thought})
        
        # 3. CHECK: Final answer?
        if is_final_answer(action):
            return action
        
        # 4. ACTING: Execute tool
        tool = tools[action]
        observation = tool.execute(action_input)  # <- Agent acts
        history.append({action: action, observation: observation})
    
    return 'Max iterations reached'
"""

print(pseudocode)
```

```python
# CELDA 7: Ventajas/Desventajas

print("\n" + "=" * 70)
print("VENTAJAS vs DESVENTAJAS")
print("=" * 70)

advantages = """
VENTAJAS:
  + Flexible: cualquier secuencia de actions
  + Adaptativo: ajusta segun observaciones
  + Transparente: vemos el reasoning
  + Poderoso: reasoning + acting
  + Escalable: agregar tools sin recodificar

DESVENTAJAS:
  - Costo: multiple LLM calls
  - Errores: pueden propagar
  - Control: unpredictable
  - Latencia: iteraciones + tools
  - Seguridad: ejecutar tools arbitrarias
"""

print(advantages)
```
