```python
# Celda 2: Frontend Setup - Next.js + React + TailwindCSS 🎨

"""
CELDA 2: Frontend Setup - Next.js Project Structure
====================================================

Crear proyecto Next.js con estructura lista para producción
"""

import os
from pathlib import Path

print("=" * 70)
print("NEXT.JS 14 PROJECT SETUP")
print("=" * 70)

print("""
¿QUÉ ES NEXT.JS?

Framework React con:
  ✅ Server-Side Rendering (SSR) - mejor SEO, performance
  ✅ Static Generation (SSG) - pre-render en build
  ✅ API Routes - backend en Node.js (opcional)
  ✅ TypeScript nativo
  ✅ Automatic code splitting
  ✅ Vercel deployment ready

Vs React puro:
  React:    SPA (Single Page App), todo en JS del cliente
  Next.js:  SSR/SSG + SPA híbrido, mejor performance

Para Churn Prediction:
  Necesitamos SSR porque queremos:
  ✅ Página carga rápido (SEO friendly)
  ✅ Datos precargados (dashboard)
  ✅ API calls asíncronos (fetch predictions)
""")

print("\n" + "=" * 70)
print("COMANDO CREAR PROYECTO")
print("=" * 70)

setup_command = """
# Terminal
npm create next-app@latest churn-prediction --typescript --tailwind

# Responder:
✓ Would you like to use ESLint? → Yes
✓ Would you like to use Tailwind CSS? → Yes
✓ Would you like to use `src/` directory? → Yes
✓ App Router? → Yes
✓ Customize import alias? → No

# Resultado: carpeta 'churn-prediction/' lista
cd churn-prediction
npm run dev
# Abre: http://localhost:3000
"""

print(setup_command)

print("\n" + "=" * 70)
print("ESTRUCTURA PROYECTO")
print("=" * 70)

structure = """
churn-prediction/
├── src/
│   ├── app/
│   │   ├── layout.tsx         ← Layout raíz
│   │   ├── page.tsx           ← Página home
│   │   ├── api/
│   │   │   └── health/
│   │   │       └── route.ts   ← API route (opcional)
│   │   └── components/
│   │       ├── PredictionForm.tsx    ← Form para input
│   │       ├── ResultsDisplay.tsx    ← Mostrar predicción
│   │       └── MetricsCard.tsx       ← KPIs dashboard
│   ├── lib/
│   │   └── api.ts             ← HTTP client (fetch)
│   └── styles/
│       └── globals.css        ← Global styles
├── public/
│   └── logo.svg
├── package.json
├── tailwind.config.ts
├── tsconfig.json
└── next.config.js
"""

print(structure)

print("\n" + "=" * 70)
print("ARCHIVOS A CREAR")
print("=" * 70)

files_to_create = {
    "src/lib/api.ts": "HTTP client (fetch con FastAPI)",
    "src/app/components/PredictionForm.tsx": "Form inputs",
    "src/app/components/ResultsDisplay.tsx": "Mostrar resultado",
    "src/app/page.tsx": "Home page completa",
}

print()
for file, purpose in files_to_create.items():
    print(f"  {file:40} → {purpose}")

print("\n" + "=" * 70)
print("SRC/LIB/API.TS - HTTP CLIENT")
print("=" * 70)

api_client = '''// src/lib/api.ts
// HTTP client para comunicarse con FastAPI backend

const API_BASE_URL = process.env.NEXT_PUBLIC_API_URL || "http://localhost:8000";

export interface PredictionRequest {
  square_feet: number;
  bedrooms: number;
  age: number;
}

export interface PredictionResponse {
  prediction: boolean;
  churn_probability: number;
  confidence: number;
  request_id: string;
  timestamp: string;
}

export async function predictChurn(
  data: PredictionRequest
): Promise<PredictionResponse> {
  const response = await fetch(`${API_BASE_URL}/predict`, {
    method: "POST",
    headers: {
      "Content-Type": "application/json",
    },
    body: JSON.stringify(data),
  });

  if (!response.ok) {
    throw new Error(`API error: ${response.status}`);
  }

  return response.json();
}

export async function getModelInfo() {
  const response = await fetch(`${API_BASE_URL}/model-info`);
  
  if (!response.ok) {
    throw new Error(`API error: ${response.status}`);
  }

  return response.json();
}

export async function getMetrics() {
  const response = await fetch(`${API_BASE_URL}/metrics`);
  
  if (!response.ok) {
    throw new Error(`API error: ${response.status}`);
  }

  return response.json();
}
'''

print(api_client)

print("\n" + "=" * 70)
print("SRC/APP/COMPONENTS/PREDICTIONFORM.TSX")
print("=" * 70)

form_component = '''// src/app/components/PredictionForm.tsx
"use client";

import { useState } from "react";
import { PredictionRequest } from "@/lib/api";

interface PredictionFormProps {
  onSubmit: (data: PredictionRequest) => void;
  isLoading: boolean;
}

export default function PredictionForm({
  onSubmit,
  isLoading,
}: PredictionFormProps) {
  const [squareFeet, setSquareFeet] = useState(2500);
  const [bedrooms, setBedrooms] = useState(3);
  const [age, setAge] = useState(10);

  const handleSubmit = (e: React.FormEvent) => {
    e.preventDefault();
    onSubmit({
      square_feet: squareFeet,
      bedrooms,
      age,
    });
  };

  return (
    <form onSubmit={handleSubmit} className="space-y-6">
      <div>
        <label className="block text-sm font-medium">
          Square Feet: {squareFeet}
        </label>
        <input
          type="range"
          min="500"
          max="10000"
          value={squareFeet}
          onChange={(e) => setSquareFeet(Number(e.target.value))}
          className="w-full"
        />
      </div>

      <div>
        <label className="block text-sm font-medium">
          Bedrooms: {bedrooms}
        </label>
        <input
          type="range"
          min="1"
          max="10"
          value={bedrooms}
          onChange={(e) => setBedrooms(Number(e.target.value))}
          className="w-full"
        />
      </div>

      <div>
        <label className="block text-sm font-medium">Age: {age}</label>
        <input
          type="range"
          min="0"
          max="150"
          value={age}
          onChange={(e) => setAge(Number(e.target.value))}
          className="w-full"
        />
      </div>

      <button
        type="submit"
        disabled={isLoading}
        className="w-full bg-blue-600 hover:bg-blue-700 text-white py-2 rounded disabled:opacity-50"
      >
        {isLoading ? "Prediciendo..." : "Predecir Churn"}
      </button>
    </form>
  );
}
'''

print(form_component)

print("\n" + "=" * 70)
print("SRC/APP/PAGE.TSX - HOME PAGE COMPLETA")
print("=" * 70)

home_page = '''// src/app/page.tsx
"use client";

import { useState } from "react";
import PredictionForm from "./components/PredictionForm";
import ResultsDisplay from "./components/ResultsDisplay";
import { predictChurn, PredictionRequest, PredictionResponse } from "@/lib/api";

export default function Home() {
  const [prediction, setPrediction] = useState<PredictionResponse | null>(null);
  const [isLoading, setIsLoading] = useState(false);
  const [error, setError] = useState<string | null>(null);

  const handlePredict = async (data: PredictionRequest) => {
    setIsLoading(true);
    setError(null);
    
    try {
      const result = await predictChurn(data);
      setPrediction(result);
    } catch (err) {
      setError(`Error: ${err instanceof Error ? err.message : "Unknown error"}`);
    } finally {
      setIsLoading(false);
    }
  };

  return (
    <div className="min-h-screen bg-gray-50">
      <div className="max-w-2xl mx-auto px-4 py-8">
        <h1 className="text-4xl font-bold mb-2">Churn Prediction</h1>
        <p className="text-gray-600 mb-8">
          Predice si un cliente tiene riesgo de abandonar el servicio
        </p>

        <div className="grid grid-cols-1 md:grid-cols-2 gap-8">
          <div className="bg-white p-6 rounded-lg shadow">
            <h2 className="text-xl font-semibold mb-4">Ingresa Datos</h2>
            <PredictionForm onSubmit={handlePredict} isLoading={isLoading} />
          </div>

          <div className="bg-white p-6 rounded-lg shadow">
            <h2 className="text-xl font-semibold mb-4">Resultado</h2>
            {error && <div className="text-red-600">{error}</div>}
            {prediction && <ResultsDisplay prediction={prediction} />}
            {!prediction && !error && (
              <p className="text-gray-400">Completa el formulario y presiona predecir</p>
            )}
          </div>
        </div>
      </div>
    </div>
  );
}
'''

print(home_page)

print("\n" + "=" * 70)
print("SIGUIENTES PASOS")
print("=" * 70)

next_steps = """
1. Crear carpeta frontend:
   $ cd projects/mes3
   $ npm create next-app@latest churn-app --typescript --tailwind

2. Copiar archivos:
   - src/lib/api.ts
   - src/app/components/PredictionForm.tsx
   - src/app/components/ResultsDisplay.tsx (en Celda 3)
   - src/app/page.tsx

3. Instalar dependencias extra (si necesario):
   $ npm install axios  # (opcional, usamos fetch)

4. Ejecutar dev server:
   $ npm run dev
   → http://localhost:3000

5. Conectar con backend:
   - Backend debe estar corriendo (docker-compose up)
   - Frontend hace fetch a http://localhost:8000
"""

print(next_steps)

print("\n✅ Celda 2: Entiendes Next.js + React + TailwindCSS")
```

```python
# Celda 3: ResultsDisplay Component 📊

"""
CELDA 3: Componente ResultsDisplay
===================================

Mostrar predicción con UI/UX profesional
"""

print("=" * 70)
print("COMPONENTE RESULTSDISPLAY")
print("=" * 70)

print("""
LÓGICA DE VISUALIZACIÓN:

Score de Churn (0-100%):
  0-30%   → 🟢 Verde: BAJO RIESGO
  30-60%  → 🟡 Amarillo: RIESGO MEDIO
  60-100% → 🔴 Rojo: ALTO RIESGO

Recomendaciones:
  - Score bajo: "No requiere acción inmediata"
  - Score medio: "Monitorear, enviar promoción"
  - Score alto: "Contactar urgente, oferta especial"
""")

print("\n" + "=" * 70)
print("SRC/APP/COMPONENTS/RESULTSDISPLAY.TSX")
print("=" * 70)

results_component = '''// src/app/components/ResultsDisplay.tsx
import { PredictionResponse } from "@/lib/api";

interface ResultsDisplayProps {
  prediction: PredictionResponse;
}

export default function ResultsDisplay({ prediction }: ResultsDisplayProps) {
  const churnProbability = prediction.churn_probability * 100;
  
  // Determinar color y mensaje según score
  const getRiskLevel = (prob: number) => {
    if (prob < 30) return { level: "BAJO", color: "bg-green-500", text: "text-green-700" };
    if (prob < 60) return { level: "MEDIO", color: "bg-yellow-500", text: "text-yellow-700" };
    return { level: "ALTO", color: "bg-red-500", text: "text-red-700" };
  };

  const getRecommendation = (prob: number): string => {
    if (prob < 30) {
      return "✅ Cliente satisfecho. Mantener relación actual.";
    }
    if (prob < 60) {
      return "⚠️ Riesgo moderado. Enviar promoción especial o descuento. Aumentar comunicación.";
    }
    return "🚨 Alto riesgo. Contactar inmediatamente. Ofrecer plan mejorado o descuento significativo.";
  };

  const riskLevel = getRiskLevel(churnProbability);
  const recommendation = getRecommendation(churnProbability);

  return (
    <div className="space-y-6">
      {/* SCORE PRINCIPAL */}
      <div className="text-center">
        <div className="text-6xl font-bold text-gray-800">
          {churnProbability.toFixed(1)}%
        </div>
        <p className="text-gray-600 mt-2">Probabilidad de Churn</p>
      </div>

      {/* BARRA DE PROGRESO COLORIDA */}
      <div className="w-full bg-gray-200 rounded-full h-8 overflow-hidden">
        <div
          className={`${riskLevel.color} h-full transition-all duration-500`}
          style={{ width: `${churnProbability}%` }}
        />
      </div>

      {/* NIVEL DE RIESGO */}
      <div className={`p-4 rounded-lg ${riskLevel.text} bg-opacity-10`}>
        <p className="text-xl font-semibold">
          Riesgo: <span className={riskLevel.text}>{riskLevel.level}</span>
        </p>
      </div>

      {/* RECOMENDACIÓN */}
      <div className="bg-blue-50 border-l-4 border-blue-500 p-4">
        <p className="text-blue-800">
          <strong>Recomendación:</strong> {recommendation}
        </p>
      </div>

      {/* CONFIANZA DEL MODELO */}
      <div className="bg-gray-100 p-3 rounded text-sm text-gray-700">
        <p>
          <strong>Confianza:</strong> {(prediction.confidence * 100).toFixed(1)}%
        </p>
        <p className="text-xs text-gray-600 mt-1">
          Request ID: {prediction.request_id}
        </p>
      </div>

      {/* METADATA */}
      <div className="text-xs text-gray-500 space-y-1">
        <p>📅 {new Date(prediction.timestamp).toLocaleString("es-ES")}</p>
      </div>
    </div>
  );
}
'''

print(results_component)

print("\n" + "=" * 70)
print("FLUJO COMPLETO: USER JOURNEY")
print("=" * 70)

user_journey = """
1. Usuario abre http://localhost:3000
   → Ve formulario con 3 sliders (square_feet, bedrooms, age)

2. Ajusta valores y presiona "Predecir Churn"
   → Frontend hace POST a http://localhost:8000/predict

3. Backend (FastAPI):
   - Valida input con Pydantic
   - Carga modelo ML entrenado
   - Hace predicción
   - Devuelve JSON: {prediction, churn_probability, confidence, ...}

4. Frontend recibe respuesta
   → Muestra ResultsDisplay con:
     - Score en grande
     - Barra de progreso colorida
     - Nivel de riesgo (BAJO/MEDIO/ALTO)
     - Recomendación específica
     - Confianza del modelo

5. Usuario ve resultado y decide qué hacer
"""

print(user_journey)

print("\n" + "=" * 70)
print("INTEGRACIÓN FRONTEND ↔ BACKEND")
print("=" * 70)

integration = """
FRONTEND (Next.js)          BACKEND (FastAPI)
────────────────────       ────────────────────

localhost:3000
    ↓
PredictionForm ─POST→ localhost:8000/predict
    ↓                        ↓
  [formulario]          [validación Pydantic]
  [sliders]             [carga modelo ML]
  [botón]               [inference]
    ↓                        ↓
ResultsDisplay ←JSON─ [resultado predicción]
  [score]
  [barra]
  [recomendación]

KEYS:
✅ CORS: FastAPI debe permitir http://localhost:3000
✅ .env: Frontend conoce URL backend
✅ Error handling: Mostrar si API falla
"""

print(integration)

print("\n" + "=" * 70)
print("CONFIGURACIÓN: CORS EN FASTAPI")
print("=" * 70)

cors_setup = '''# En main.py del backend FastAPI

from fastapi.middleware.cors import CORSMiddleware

app = FastAPI()

app.add_middleware(
    CORSMiddleware,
    allow_origins=["http://localhost:3000", "http://localhost:8000"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

# Ahora frontend puede hacer fetch sin problemas
'''

print(cors_setup)

print("\n" + "=" * 70)
print("ARCHIVO .ENV.LOCAL (FRONTEND)")
print("=" * 70)

env_file = """
# .env.local en raíz del proyecto Next.js
NEXT_PUBLIC_API_URL=http://localhost:8000

(public = accesible desde cliente)
(privado = solo en servidor)
"""

print(env_file)

print("\n" + "=" * 70)
print("TESTING DE COMPONENTES")
print("=" * 70)

testing = """
Cómo probar manualmente:

1. Terminal 1: Levantar backend
   $ cd ~/ia-master
   $ docker-compose up

2. Terminal 2: Frontend dev server
   $ cd projects/mes3/churn-app
   $ npm run dev

3. Browser: http://localhost:3000
   - Ajusta sliders
   - Presiona "Predecir"
   - Observa resultado

Debugging:
  - DevTools (F12) → Network → ver request/response
  - Console → ver errores de fetch
  - Backend logs → docker-compose logs api
"""

print(testing)

print("\n" + "=" * 70)
print("SIGUIENTES PASOS CELDA 4")
print("=" * 70)

next = """
Celda 4: Testing End-to-End (E2E)
  - Simular clicks en formulario
  - Validar que respuesta se muestra
  - Verificar error handling

Herramientas:
  - Playwright (recomendado)
  - Cypress
  - Selenium
"""

print(next)

print("\n✅ Celda 3: Creaste componente de resultados profesional")
```
